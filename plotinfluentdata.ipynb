```python
import pandas as pd
import matplotlib.pyplot as plt

# Bestandsnamen (pas aan naar jouw bestandsnamen)
filenames = [
    "TKN_cycli.csv",
    "TSS_interpolated.csv",
    "CODt_interpolated.csv",
    "flowrate_interpolated.csv",
    "temperature_interpolated.csv",
    "DObasin4_interpolated.csv"
]

# Itereer over elk bestand en plot
for i, fname in enumerate(filenames):
    df = pd.read_csv(fname, parse_dates=["Datetime"])

    plt.figure(figsize=(12, 4))
    plt.plot(df["Datetime"], df[df.columns[1]], marker='o', linestyle='-', markersize=2, label=f"Bestand {i+1}")
    plt.xlabel("Time")
    plt.ylabel(df.columns[1])
    plt.title(f"Values  Deel {i+1}")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
